```python
# plot sinx,cosx in same plot
x = np.arange(-m.pi, m.pi, 0.1)
y = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y, color = 'r',linewidth=2.0, linestyle="-")
plt.plot(x, y2, color = 'g',linewidth=2.0, linestyle="--")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Sin(x) and Cos(x) in same plot")
plt.legend(["sin(x)","cos(x)"])   # calling legend
plt.show()
```

```python
# plot sinx,cosx in same plot
x = np.arange(-m.pi, m.pi, 0.1)
y = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y, color = 'r',linewidth=2.0, linestyle="-",label = "Sin(x)")
plt.plot(x, y2, color = 'g',linewidth=2.0, linestyle="--",label = "Cos(x)")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Sin(x) and Cos(x) in same plot")
plt.legend()  # when we pass label we don't need to write anything in legend
plt.show()
```

```python
#pie chart
sizes = [2,4,6,7,10]
labels = ['a','b','c','d','e']
colors = ['yellowgreen','gold','skyblue','lightcoral','red']
plt.title("Split among classes")
explode = [0.2,0.1,0.1,0.1,0.1]  # explodes the graph in different parts
plt.axis('equal')
plt.pie(sizes,colors=colors,explode=explode,labels=labels,autopct='%.1f%%',shadow=True,startangle=90)
plt.show()
```
