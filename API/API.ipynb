```python
from skimage import io
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = [5.0,5.0]
plt.rcParams["figure.autolayout"] = True
for i in range(0,5):
    response4 = rq.get('https://dog.ceo/api/breeds/image/random')
    message = response4.json()['message']
    f = message
    a = io.imread(f)
    plt.imshow(a)
    plt.axis('off')
    plt.show()

```

```python
# post method
p = rq.post('http://httpbin.org/post')
p.status_code
p
p.text
```

```python
# in post method we actually change(write) the data while in get method we just get(read) the data
p = rq.post('http://httpbin.org/post',data={"key1":"value1","key2":"value2"})
p.status_code
print(p.json())
```

```python
# initially this site have 100 posts
# we create new post with id = 101
r = rq.post('https://jsonplaceholder.typicode.com/posts',data={"id":101,"title":"new title","body":"new body"})
r.status_code
print(r.json())
print(r.url)
```
